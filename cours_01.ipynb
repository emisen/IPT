# I. Le Tableau

### I.1. Définition et propriétés

Le tableau est la seule structure de données présentée en première année.


Un tableau est une structure de données de même type (python autorise tout de même les tableaux dits hétérogènes c'est à dire d'éléments de type différents), modifiable, finie (les tableaux sont dynamiques en python) et telle que l’accès à un de ses éléments se fait en temps constant (on obtient la valeur d’indice i aussi rapidement peu importe la valeur de i)


Si les éléments du tableau sont positionnés en mémoire les uns à côté des autres à partir d’une adresse mémoire $m_0$ et que chaque élément du tableau prend la même place $m$ (d'où les éléments de même type), on trouve l’élément d’indice i à la position $m_0 + i\times m$ aussi rapidement **peu importe i**.


```python
import time

#on crée un tableau de 1 000 000 d'éléments
a=[2]*1000000

# on mesure 10fois 1000000 d'appels à la première valeur
t=time.time()
for j in range(10):
    for i in range(1000000):
        a[0]
t=time.time()-t
print(t)

# on mesure 10 fois 1 appel à chacune des 1000000 de valeurs
t=time.time()
for j in range(10):
    for i in range(1000000):
        a[i]
t=time.time()-t
print(t)

# on mesure 10fois 1000000 d'appels à la dernière valeur
t=time.time()
for j in range(10):
    for i in range(1000000):
        a[-1]
t=time.time()-t
print(t)


# on mesure 10fois 1000000 d'appels à une valeur du tableau
t=time.time()
for j in range(10):
    for i in range(1000000):
        a[324598]
t=time.time()-t
print(t)
```

Chaque case du tableau garde son adresse mémoire mais est modifiable (la valeur stockée à cette adresse mémoire peut être changée)


```python
# Création d'un tableau de 4 éléments
a = [1.27,2.45,3.798,4.142]

# Affichage du tableau et de l'adresse mémoire de a
print(a,hex(id(a)))

# modification d'un élément
a[2]=1.182

# Affichage du tableau et de l'adresse mémoire de a
print(a,hex(id(a)))
```

On voit bien que l'adresse mémoire de **a** n'a pas changé ici.

### I.2. Que peut-on faire avec un objet de type *list* en python ?

* Créer une liste vide

```python
t = []
```

* ajouter un élément à la fin

* Ou à la position que l'on souhaite

```python
t = [1,2,3,4,2,3,2,2,7]

print(t.index(3))
print(t.count(2))
t.sort()
print(t)
print(len(t))
```

### Attention toutefois à la copie d'objets de type **list** car python est fait pour limiter l'usage de la mémoire ainsi:

```python
t = [1, 2, 3, 4]
s = t # ne copie pas t. On a 1 seul objet référencé deux fois

# Pour preuve, l'adresse mémoire est la même
print(hex(id(t)),hex(id(s)))
```

```python
t[0] = 4
print(s,t) # Les deux sont modifiées par la ligne précédente
```

Une solution est de forcer la copie avec la syntaxe des slices.

```python
t = [1, 2, 3, 4]
s = t[ : ]

# adresses mémoire différentes
print(hex(id(t)),hex(id(s)))

t[0] = 4
print(s,t) # seule t est modifiée
```

Mais il y a encore plus perturbant, on peut **copier une *list* sans copier les sous *list***

```python
t = [[1,2,3], [4,5,6], [7,8,9]]
s = t[:]

# adresses mémoire différentes
print(hex(id(t)),hex(id(s)))

# mais chaque sous tableau n'existe qu'une seule fois en mémoire
print(hex(id(t[0])),hex(id(s[0])))
```

Ainsi, si vous travaillez avec des matrices, il faudra utiliser la fonction **deepcopy** de la bibliothèque **copy** pour les copier afin d'éviter tout problème. 

```python
t = [[1,2,3], [4,5,6], [7,8,9]]
s = t[:]

# pas de problème on modifie s[0] qui est bien différent de t[0] puisque s et t sont des objets différents en mémoire
s[0] = 0
print(s, t)
```

```python
t = [[1,2,3], [4,5,6], [7,8,9]]
s = t[:]

# problème on modifie s[0][0] qui est exactement le même que t[0][0], les deux listes sont modifiées
s[0][0] = 0
print(s, t)
```

```python
from copy import deepcopy
t = [[1,2,3], [4,5,6], [7,8,9]]
s = deepcopy(t)

# Cette fois ci pas de problème, deepcopy a fait une copie de l'ensemble des données
s[0] = 0
print(s, t)
```

### I.3. Avec numpy

Contrairement aux ***list*** de python, on a ici des tableaux qui sont **homogènes** (un seul type de données) et qui sont **statiques**. Il faudra donc définir à l'avance la taille de ces objets et le type des données contenues. Si le type n'est pas précisé, numpy le "devine" à partir des données fournies.
Un petit exemple avec des entiers:

```python
import numpy as np
vect1 = np.array([1, 2, 3])
print(vect1)

vect2 = np.array([1, 2, 3], dtype=np.float)
print(vect2)
```

Que se passe-t-il si on essaye de mettre autre chose dans ce tableau ?

```python
vect1[1] = 3.7
print(vect1)

vect2[1] = 3.7
print(vect2)

vect1[2]="hello"
print(vect1)
```

Soit une modification est possible sur la donnée (ici le float a été tronqué en int) et on conserve un tableau avec un seul type de données (attention la donnée a été modifiée sans avertissement!) soit on reçoit un message d'erreur si cela n'est pas possible (cas de la chaine de caractère)

Il faut donc bien faire la distinction entre les tableau "python" et les tableaux "numpy" qui n'ont pas les mêmes propriétés exactement car nous manipulerons les deux cette année.
