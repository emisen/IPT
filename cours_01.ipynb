Une structure de données est l’objet informatique qui stocke des valeurs en mémoire. Nous allons ici n'en étudier que certaines. Pour caractériser une structure de données, il faut connaître les opérations possibles sur celle-ci et leur coût.


# I. Le Tableau

Le tableau est la seule structure de données présentée en première année.


Un tableau est une structure de données de même type (python autorise tout de même les tableaux dits hétérogènes c'est à dire d'éléments de type différents), modifiable, finie (les tableaux sont dynamiques en python) et telle que l’accès à un de ses éléments se fait en temps constant (on obtient la valeur d’indice i aussi rapidement peu importe la valeur de i)


Si les éléments du tableau sont positionnés en mémoire les uns à côté des autres à partir d’une adresse mémoire $m_0$ et que chaque élément du tableau prend la même place $m$ (d'où les éléments de même type), on trouve l’élément d’indice i à la position $m_0 + i\times m$ aussi rapidement **peu importe i**.


```python
import time

#on crée un tableau de 1 000 000 d'éléments
a=[2]*1000000

# on mesure 10fois 1000000 d'appels à la première valeur
t=time.time()
for j in range(10):
    for i in range(1000000):
        a[0]
t=time.time()-t
print(t)

# on mesure 10 fois 1 appel à chacune des 1000000 de valeurs
t=time.time()
for j in range(10):
    for i in range(1000000):
        a[i]
t=time.time()-t
print(t)

# on mesure 10fois 1000000 d'appels à la dernière valeur
t=time.time()
for j in range(10):
    for i in range(1000000):
        a[-1]
t=time.time()-t
print(t)


# on mesure 10fois 1000000 d'appels à une valeur du tableau
t=time.time()
for j in range(10):
    for i in range(1000000):
        a[324598]
t=time.time()-t
print(t)
```

Chaque case du tableau garde son adresse mémoire mais est modifiable (la valeur stockée à cette adresse mémoire peut être changée)


```python
# Création d'un tableau de 4 éléments
a = [1.27,2.45,3.798,4.142]

# Affichage du tableau et de l'adresse mémoire de a
print(a,hex(id(a)))

# modification d'un élément
a[2]=1.182

# Affichage du tableau et de l'adresse mémoire de a
print(a,hex(id(a)))
```

On voit bien que l'adresse mémoire de **a** n'a pas changé ici.

### Que peut-on faire avec un objet de type *list* en python ?

* Créer une liste vide

```python
t = []
```

* ajouter un élément à la fin

```python
t.append(3)
print(t)

t.append(6)
print(t)
```

* Ou à la position que l'on souhaite

```python
t.insert(1,5)
print(t)
```

* On peut écrire une valeur dans une case de la liste ou la lire

```python
element = t[1]
print(element , t)

t[2] = 4
print(element , t)
```

On rappelle que les indices commencent à $0$ et vont jusqu'à $n-1$ en python pour une list de n élément.


La méthode **pop** permet de renvoyer un élément (par défaut d'argument, le dernier de la liste) et de le supprimer de la liste.

```python
print(t.pop(),t)

print(t.pop(),t)
```

Les méthodes **index(x)**, **count(x)** et **sort** permettent respectivement de renvoyer l'indice de la première occurence de x, le nombre d'occurences de x et de trier la liste.
Enfin **len(t)** renvoie le nombre d'éléments de t

```python
t = [1,2,3,4,2,3,2,2,7]

print(t.index(3))
print(t.count(2))
print(len(t))
t.sort()
print(t)
```

On peut aussi utiliser des indices négatifs qui permettent de pointer les éléments depuis la fin

```python
t[-1]
```

Ou utiliser les slices:
* **t[ i : j ]** renvoie les éléments d'indice i à j-1
* **t[ : j ]** du début jusqu'à j-1
* **t[ i : ]** de l'indice i jusqu'à la fin

```python
print(t[1:6],t[:8],t[6:])
```

Enfin t[:] peut être utilisé pour faire une copie d'une liste, et un troisième indice permet de filtrer en imposant un pas.

```python
s = t[:]
print(t,s)
```

```python
t[1:6:2] # les élément d'indices 1 à 5 incluspar pas de 2
```

```python
t[: : -1] # Les éléments de la fin jusqu'au début par pas de -1
```
