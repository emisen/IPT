## II. Pile d'appel

Lorsque l'on lance une fonction récursive, elle va rester en attente du résultat d'un appel à elle même. Cet état est stocké dans une structure de pile avec toutes les variables locales associées à cet état. Si le nouvel appel a besoin d'appeler une nouvelle fois la fonction, son état est stocké au sommet de la pile et ainsi de suite.
Considérons la fonction suivante qui a pour but de calculer $x^n$

```python
def puiss(x,n):
    assert type(x)==float or type(x)==int
    assert type(n)==int
    if n == 0:
        return 1
    else:
        reste = puiss(x,n-1)
        return x * reste
```

Imaginons que l'on lance **puiss(2,3)**

on a à ce niveau **x = 2**, et **n == 3**. On rentre donc dans la branche **else** et on se retrouve bloqué lors de l'afectation de la variable **reste**. On stocke alors tout ça dans une pile dont l'état est:

| Etat | x | n | reste |
| :---: | :---: | :---: | :---: |
| puiss(2,3) | 2 | 3 | ? |


On appelle **puiss(2,2)**, puis **puiss(2,1)**. Le nouvel état de la pile est:


| Etat | x | n | reste |
| :---: | :---: | :---: | :---: |
| puiss(2,1) | 2 | 1 | ? |
| puiss(2,2) | 2 | 2 | ? |
| puiss(2,3) | 2 | 3 | ? |

Chaque variable est locale, elle est différente de celle du niveau suivant ou précédent dans la pile. On appelle alors **puiss(2,0)** pour laquelle on rentre dans la branche **if** et qui renvoie **1**. Ce résultat est donc renvoyé au niveau su dessous dans la pile d'appel

| Etat | x | n | reste |
| :---: | :---: | :---: | :---: |
| puiss(2,1) | 2 | 1 | 1 |
| puiss(2,2) | 2 | 2 | ? |
| puiss(2,3) | 2 | 3 | ? |

Ainsi **puiss(2,1)** peut être calculé, l'élément du dessus de la pile est dépilé et le résultat est renvoyé au niveau suivant.

| Etat | x | n | reste |
| :---: | :---: | :---: | :---: |
| puiss(2,2) | 2 | 2 | 2 |
| puiss(2,3) | 2 | 3 | ? |

Et ainsi de suite jusqu'à ce que la pile soit vide et que l'on obtienne le résultat.

Tout se passe comme si on calculait:
$2^3 = 2 * (2^2) = 2 * (2 * (2^1)) = 2 * (2 * (2 * (2^0))) = 2 * (2 * (2 * (1))) = 2 * (2 * (2)) = 2 * (4) = 8$


Tout ceci coûte donc beaucoup de mémoire car on est obligé de stocker la fonction et l'état de ses arguments à chaque étape et on ne peut pas faire ça à l'infini.

La **récursivité terminale** permet d'améliorer les choses car elle permet d'obtenir le résultat souhaité avant d'avoir dépilé toute la pile d'appel. Par exemple, pour la fonction puissance précédente, on peut stocker le résultat précédent et le passer au niveau du dessus:

```python
def puiss2(x,n,a=1):
    assert type(x)==float or type(x)==int
    assert type(n)==int
    if n == 0:
        return a
    else:
        return puiss2(x,n-1,a*x)
```

En python, une fonction récursive est plus lente que son homologue itérative et consomme plus de place en mémoire. Il ne faudra donc l'utiliser que si cela simplifie l'écriture du programme ou la lisibilité. De plus aucune optimisation ne permet de profiter du concept de récursivité terminale, la pile d'appel sera donc dépilée en entier même si ce n'est pas nécessaire. 

D'autres langages de programmation compilés permettent d'obtenir de bons résultats en terme de temps d'exécution avec des fonctions récursives et récursives terminales à condition que le compilateur soit optimisé pour cela.

### Comparaison des fonctions puissance récursive, récursive terminale et itérative en temps d'exécution

```python
def puiss(x,n):
    assert type(x)==float or type(x)==int
    assert type(n)==int
    if n == 0:
        return 1
    else:
        return x * puiss(x,n-1)
    
def puiss2(x,n,a=1):
    assert type(x)==float or type(x)==int
    assert type(n)==int
    if n == 0:
        return a
    else:
        return puiss2(x,n-1,a*x)
    
def puiss3(x,n):
    reponse = 1
    for i in range(n):
        reponse = reponse * x
    return reponse
```

```python
from time import time

t = time()
for i in range(1000):
    puiss(2,1000)
print(time()-t)

t = time()
for i in range(1000):
    puiss2(2,1000)
print(time()-t)

t = time()
for i in range(1000):
    puiss3(2,1000)
print(time()-t)
```

En python, la fonction itérative est presque 10 fois plus rapide...
