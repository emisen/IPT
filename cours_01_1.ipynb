# II. La Pile

### II.1. Définition et propriétés



Il faut imaginer que les données sont empilées comme dans une pile d'assiettes.



La première donnée arrivée sera en dessous de la pile puis la suivante juste dessus etc…



A chaque instant, on n’a accès directement qu’à l’assiette qui est au sommet de la pile



On appelle cette structure une **pile** ou **stack** en anglais. On parle encore de **LIFO** pour last in first out


Pour la structure de pile il faut pouvoir:
* créer une pile vide (de taille définie) => init
* vérifier si la pile est vide => isEmpty
* empiler un élément => push
* dépiler un élément et le renvoyer => pop
    
**Chacune de ces opérations se fait en temps constant.**

```python
def init(n):
    """
    Crée une pile de taille n
    On utilise ici une list de taille n+1
    la première case indique le nombre d'éléments
    les autres contiennent les données (ou pas)
    """
    return [0]+[None]*n

def isEmpty(stack):
    return stack[0] == 0

def push(element, stack):
    assert stack[0] < len(stack)-1
    stack[0] += 1
    stack[stack[0]] = element
    
def pop(stack):
    val = stack[stack[0]]
    stack[0] -= 1
    return val
```
